## Data prep

Prep data that was manually transposed from the Excel file to be Ayasdi ready.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/KEGGs-8-zeros-filtered-transposed.csv")
```

```python
df.head()
```

```python
df = df.iloc[:,1:]
df.head()
```

```python
keggs = df.columns[1:]
```

```python
keggs_simple = [kegg.split("(")[0] for kegg in keggs]
```

```python
df.columns = ["Subject"] +  keggs_simple
```

```python
df.head()
```

```python
df.Subject
```

```python
df["Cohort"]= df.Subject.apply(lambda x: x[0:2])
```

```python
df.Cohort
```

```python
df.to_csv("../data/KEGGs-8-zeros-filtered-transposed.csv")
```
